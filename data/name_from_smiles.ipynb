```python
from pubchempy import get_compounds
import pandas as pd
```

```python
smiles_list = pd.read_csv('GABAA_antagonist.csv')['SMILES']
```

```python
smiles_list
```

```python
iupac_name_list = []
```

```python
for smiles in smiles_list:
    compounds = get_compounds(smiles, 'smiles')
    for compound in compounds:
        iupac_name_list.append(compound.iupac_name)
        print(compound.iupac_name)
```

```python
len(iupac_name_list)
```

```python
df = pd.DataFrame({'SMILES': smiles_list, 'IUPAC Name': iupac_name_list})
```

```python
df
```

```python
df.to_csv('GABAA_antagonist.csv')
```
