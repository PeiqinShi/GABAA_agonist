# Similarity

```python
import pandas as pd
from rdkit import Chem
from rdkit.DataStructs import FingerprintSimilarity
```

```python
data = pd.read_csv('GABAA.csv',encoding='gb18030')
plant = pd.read_csv('plant_data.csv',encoding='gb18030')
```

```python
# Define SMILES for dataset 1
dataset1_smiles = data['smiles'][0:244]

# Define SMILES for dataset 2
dataset2_smiles = plant['smiles']

# creat DataFrame of result 
AD_results = pd.DataFrame(columns=['SMILES', 'InDomain'])

# Calculate the similarity of each molecule in dataset 2 to dataset 1 and determine whether it is in the application domain
for smiles in dataset2_smiles:
    # Convert to RDKit molecular objects
    mol2 = Chem.MolFromSmiles(smiles)
    
    # Initializes the application domain flag
    in_domain = False
    
    # Calculate the similarity to each molecule in dataset 1
    for ref_smiles in dataset1_smiles:
        # Convert to RDKit molecular objects
        mol1 = Chem.MolFromSmiles(ref_smiles)
        
        # Calculate Tanimoto similarity
        similarity = FingerprintSimilarity(Chem.RDKFingerprint(mol1), Chem.RDKFingerprint(mol2))
        #print(f"Tanimoto similarity between {smiles} and {ref_smiles}: {similarity}")
        # Check whether it is in the application domain
        if similarity > 0.6: 
            in_domain = True
    
    # Add result to DataFrame
    AD_results = AD_results._append({'SMILES': smiles, 'InDomain': in_domain}, ignore_index=True)
```

```python
AD_results
```

```python
# Calculate the number of True and False
value_counts = AD_results['InDomain'].value_counts()


print(value_counts)
```

```python
AD_results.to_csv('AD_results_final.csv')
```
